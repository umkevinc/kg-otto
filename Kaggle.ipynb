```python
from itertools import combinations
def feature_gen_square(X):
    for col_name in X.columns:
        X['%s_sqr' % col_name] = X[col_name]*X[col_name]

def feature_gen_log(X):
    for col_name in X.columns:
        X['%s_log' % col_name] = np.log(X[col_name]+1)
       
def all_combinations(X):
    for cb in combinations(X.columns, 2):
        X['%s_%s' % cb] = X[cb[0]] * X[cb[1]]

#all_combinations(X)
feature_gen_square(X)
feature_gen_log(X)

```

```python
from sklearn.ensemble import RandomForestClassifier
def random_forest_search(n, X_train_minmax, y_train, X_test_minmax, y_test, **kwargs):
    clf = RandomForestClassifier(n_estimators=n, n_jobs=4, **kwargs)
    clf.fit(X_train_minmax, y_train)
    train_error = (1-clf.score(X_train_minmax, y_train))
    test_error = (1-clf.score(X_test_minmax, y_test))
    train_logloss = log_loss(y_train, clf.predict_proba(X_train_minmax))
    test_log_loss = log_loss(y_test, clf.predict_proba(X_test_minmax))
    print 'n_estimators: %s' % n
    print 'Training Error: %s' % train_error
    print 'Training Error: %s' % test_error
    print 'train log_loss: %s' % train_logloss
    print 'test log_loss: %s' % test_log_loss
    print 
    return train_error, test_error, train_logloss, test_log_loss
```

```python
def random_forest_pca_search(n, pca, X_train_minmax, y_train, X_test_minmax, y_test):
    clf = RandomForestClassifier(n_estimators=n)
    pca.fit(X_train_minmax)
    X_train_minmax = pca.transform(X_train_minmax)
    X_test_minmax = pca.transform(X_test_minmax)
    clf.fit(X_train_minmax, y_train)
    train_error = (1-clf.score(X_train_minmax, y_train))
    test_error = (1-clf.score(X_test_minmax, y_test))
    train_logloss = log_loss(y_train, clf.predict_proba(X_train_minmax))
    test_log_loss = log_loss(y_test, clf.predict_proba(X_test_minmax))
    print 'n_estimators: %s' % n
    print 'Training Error: %s' % train_error
    print 'Training Error: %s' % test_error
    print 'train log_loss: %s' % train_logloss
    print 'test log_loss: %s' % test_log_loss
    print 
    return train_error, test_error, train_logloss, test_log_loss
```

```python
from sklearn import linear_model, decomposition
from sklearn.pipeline import Pipeline
from sklearn.grid_search import GridSearchCV

#logistic = linear_model.LogisticRegression()
pca = decomposition.PCA()
pipe = Pipeline(steps=[('pca', pca), ('sgd', linear_model.SGDClassifier())])

###############################################################################
# Plot the PCA spectrum
pca.fit(X_train)
plt.figure(1, figsize=(4, 3))
plt.clf()
plt.axes([.2, .2, .7, .7])
plt.plot(pca.explained_variance_, linewidth=2)
plt.axis('tight')
plt.xlabel('n_components')
plt.ylabel('explained_variance_')

###############################################################################
# # Prediction
# n_estimators = range(10, 80, 5)
n_components = range(5,93,2)
n_search_space = range(10,200,5)
n_search_space = range(10,50,5)
#error_lst = [random_forest_pca_search(n, pca, X_train_minmax, y_train, X_test_minmax, y_test) for n in n_search_space]
# #Parameters of pipelines can be set using ‘__’ separated parameter names:

estimator = GridSearchCV(pipe,
                         dict(pca__n_components=n_components))
estimator.fit(X_train_minmax, y_train)

train_error = (1-estimator.score(X_train_minmax, y_train))
test_error = (1-estimator.score(X_test_minmax, y_test))
train_logloss = log_loss(y_train, estimator.decision_function(X_train_minmax))
test_log_loss = log_loss(y_test, estimator.decision_function(X_test_minmax))
print 'n_components: %s' % estimator.best_estimator_.named_steps['pca'].n_components
print 'Training Error: %s' % train_error
print 'Training Error: %s' % test_error
print 'train log_loss: %s' % train_logloss
print 'test log_loss: %s' % test_log_loss
print 
plt.axvline(estimator.best_estimator_.named_steps['pca'].n_components,
            linestyle=':', label='n_components chosen')
plt.legend(prop=dict(size=12))
```
